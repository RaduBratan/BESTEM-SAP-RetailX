```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json

df = pd.read_excel('sales_and_eodStocks.xlsx', sheet_name = 'Sheet1')

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

df['Month'] = df['Date'].dt.month

# Define a function to determine the season based on the month
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Autumn'
    else:
        return 'Winter'
    
def get_Christmas(date):
    if date.month == 12 and 17 <= date.day <= 24:
        return 'Christmas'
    else:
        return 'Not Christmas'
    
def get_Revelion(date):
    if date.month == 12 and 27 <= date.day <= 31:
        return 'Revelion'
    else:
        return 'Not Revelion'

def get_Dragobete(date):
    if date.month == 2 and 7 <= date.day <= 24:
        return 'Dragobete'
    else:
        return 'Not Dragobete'

def get_Halloween(date):
    if date.month == 10 and 25 <= date.day <= 31:
        return 'Halloween'
    else:
        return 'Not Halloween'
    
def get_1June(date):
    if (date.month == 5 and 28 <= date.day <= 31) or (date.month == 6 and 1 == date.day):
        return 'Child Day'
    else:
        return 'Not Child Day'
    
def get_Paste(date):
    if(date.month == 3 and 29 <= date.day <= 31) or date.month == 4 or (date.month == 5 and 1 <= date.day <= 5):
        return 'Paste'
    else:
        return 'Not Paste'
    
# Apply the function to create a 'Season' column
df['Season'] = df['Month'].apply(get_season)
df['Christmas'] = df['Date'].apply(get_Christmas)
df['Revelion'] = df['Date'].apply(get_Revelion)
df['Dragobete'] = df['Date'].apply(get_Dragobete)
df['Halloween'] = df['Date'].apply(get_Halloween)
df['Child_Day'] = df['Date'].apply(get_1June)
df['Paste_Days'] = df['Date'].apply(get_Paste)

season_sales = df.groupby(['Product_ID', 'Season'])['Sales'].sum().reset_index()
Christmas_sales = df.groupby(['Product_ID', 'Christmas'])['Sales'].sum().reset_index()
revelion_sales = df.groupby(['Product_ID', 'Revelion'])['Sales'].sum().reset_index()
dragobete_sales = df.groupby(['Product_ID', 'Dragobete'])['Sales'].sum().reset_index()
halloween_sales = df.groupby(['Product_ID', 'Halloween'])['Sales'].sum().reset_index()
childDay_sales = df.groupby(['Product_ID', 'Child_Day'])['Sales'].sum().reset_index()
paste_sales = df.groupby(['Product_ID', 'Paste_Days'])['Sales'].sum().reset_index()

winter_sales = season_sales[season_sales['Season'] == 'Winter']
winter_spring_sales_sorted = winter_sales.sort_values(by='Sales', ascending=False)
winter_sales_sorted['Product_ID'] = winter_sales_sorted['Product_ID'].astype(str)
winter_sales_sorted['ID'] = winter_sales_sorted.reset_index().index + 1

spring_sales = season_sales[season_sales['Season'] == 'Spring']
spring_sales_sorted = spring_sales.sort_values(by='Sales', ascending=False)
spring_sales_sorted['Product_ID'] = spring_sales_sorted['Product_ID'].astype(str)
spring_sales_sorted['ID'] = spring_sales_sorted.reset_index().index + 1

summer_sales = season_sales[season_sales['Season'] == 'Summer']
summer_sales_sorted = summer_sales.sort_values(by='Sales', ascending=False)
summer_sales_sorted['Product_ID'] = summer_sales_sorted['Product_ID'].astype(str)
summer_sales_sorted['ID'] = summer_sales_sorted.reset_index().index + 1


autumn_sales = season_sales[season_sales['Season'] == 'Autumn']
autumn_sales_sorted = autumn_sales.sort_values(by='Sales', ascending=False)
autumn_sales_sorted['Product_ID'] = autumn_sales_sorted['Product_ID'].astype(str)
autumn_sales_sorted['ID'] = autumn_sales_sorted.reset_index().index + 1

Christmas_sales = Christmas_sales[Christmas_sales['Christmas'] == 'Christmas']
Christmas_sales_sorted = Christmas_sales.sort_values(by='Sales', ascending=False)
Christmas_sales_sorted['Product_ID'] = Christmas_sales_sorted['Product_ID'].astype(str)
Christmas_sales_sorted['ID'] = Christmas_sales_sorted.reset_index().index + 1

Revelion_sales = Revelion_sales[Revelion_sales['Revelion'] == 'Revelion']
Revelion_sales_sorted = Revelion_sales.sort_values(by='Sales', ascending=False)
Revelion_sales_sorted['Product_ID'] = Revelion_sales_sorted['Product_ID'].astype(str)
Revelion_sales_sorted['ID'] = Revelion_sales_sorted.reset_index().index + 1

dragobete_sales = dragobete_sales[dragobete_sales['Dragobete'] == 'Dragobete']
dragobete_sales_sorted = dragobete_sales.sort_values(by='Sales', ascending=False)
dragobete_sales_sorted['Product_ID'] = dragobete_sales_sorted['Product_ID'].astype(str)
dragobete_sales_sorted['ID'] = dragobete_sales_sorted.reset_index().index + 1

halloween_sales = halloween_sales[halloween_sales['Halloween'] == 'Halloween']
halloween_sales_sorted = halloween_sales.sort_values(by='Sales', ascending=False)
halloween_sales_sorted['Product_ID'] = halloween_sales_sorted['Product_ID'].astype(str)
halloween_sales_sorted['ID'] = halloween_sales_sorted.reset_index().index + 1

childDay_sales = childDay_sales[childDay_sales['Child_Day'] == 'Child Day']
childDay_sales_sorted = childDay_sales.sort_values(by='Sales', ascending=False)
childDay_sales_sorted['Product_ID'] = childDay_sales_sorted['Product_ID'].astype(str)
childDay_sales_sorted['ID'] = childDay_sales_sorted.reset_index().index + 1

paste_sales = paste_sales[paste_sales['Paste_Days'] == 'Paste']
paste_sales_sorted = paste_sales.sort_values(by='Sales', ascending=False)
paste_sales_sorted['Product_ID'] = paste_sales_sorted['Product_ID'].astype(str)
paste_sales_sorted['ID'] = paste_sales_sorted.reset_index().index + 1

top_10_records_winter = winter_sales_sorted.head(10)
top_10_records_spring = spring_sales_sorted.head(10)
top_10_records_summer = summer_sales_sorted.head(10)
top_10_records_autumn = autumn_sales_sorted.head(10)
top_10_records_Christmas = Christmas_sales_sorted.head(10)
top_10_records_Revelion = Revelion_sales_sorted.head(10)
top_10_records_dragobete = dragobete_sales_sorted.head(10)
top_10_records_halloween = halloween_sales_sorted.head(10)
top_10_records_childDay = childDay_sales_sorted.head(10)
top_10_records_paste = paste_sales_sorted.head(10)

top_10_records_paste[['ID', 'Product_ID']].to_json('top_10_paste_bestProducts.json', orient='records', lines=True)
top_10_records_winter[['ID', 'Product_ID']].to_json('top_10_winter_bestProducts.json', orient='records', lines=True)
top_10_records_spring[['ID', 'Product_ID']].to_json('top_10_spring_bestProducts.json', orient='records', lines=True)
top_10_records_summer[['ID', 'Product_ID']].to_json('top_10_summer_bestProducts.json', orient='records', lines=True)
top_10_records_autumn[['ID', 'Product_ID']].to_json('top_10_autumn_bestProducts.json', orient='records', lines=True)
top_10_records_Christmas[['ID', 'Product_ID']].to_json('top_10_Christmas_bestProducts.json', orient='records', lines=True)
top_10_records_Revelion[['ID', 'Product_ID']].to_json('top_10_Revelion_bestProducts.json', orient='records', lines=True)
top_10_records_dragobete[['ID', 'Product_ID']].to_json('top_10_Dragobete_bestProducts.json', orient='records', lines=True)
top_10_records_halloween[['ID', 'Product_ID']].to_json('top_10_Halloween_bestProducts.json', orient='records', lines=True)
top_10_records_childDay[['ID', 'Product_ID']].to_json('top_10_childDay_bestProducts.json', orient='records', lines=True)
#apoi hardcodare un top_10_general.json
```
